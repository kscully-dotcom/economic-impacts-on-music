```python
import pandas as pd

billboard = pd.read_csv('billboard-df.csv')
playback = pd.read_csv('playbackfm-1920-2016.csv')
gdp = pd.read_csv('GDPC1.csv')
```

```python
billboard.shape
```

```python
playback.shape
```

```python
billboard['date'] = pd.to_datetime(billboard['date'], format='%Y-%m')
playback['year'] = pd.to_datetime(playback['year'], format='%Y')
gdp['DATE'] = pd.to_datetime(gdp['DATE'], format='%Y-%m')
```

```python
billboard['date'] = pd.PeriodIndex(billboard['date'], freq='Q')
playback['year'] = pd.PeriodIndex(playback['year'], freq='Q')
gdp['DATE'] = pd.PeriodIndex(gdp['DATE'], freq='Q')
```

```python
playback.rename(columns={"track name":"track_title", "track artist":"track_artist","track rank":"track_rank", "year":"date"}, inplace=True)
gdp.rename(columns={"DATE":"date"}, inplace=True)
```

```python
result = billboard.append(playback, ignore_index=True, sort=True)
result
```

```python
df = pd.merge(left=result, left_on='date', right=gdp, right_on='date', how='left')
```

```python
df
```

```python
df.to_csv('merge-billboard-gdp-df.csv', index=False)
```
